"Approved" if their "math_grade" is 70 or above,

"Rejected" if their "math_grade" is below 70.

students = [

    {"name": "ram", "math_grade": 43},

]

8.  You have two lists of numbers, and you need to find out which numbers appear in both lists. 

Given two lists of numbers [1,2,3,4,5] and [3,4,5,6,7] write a for  loop to find the common elements.

```python
list1 = [1,2,3,4,5]
list2 = [3,4,5,6,7]
list3=[]
for i in list1:
    for j in list2:
        if i==j:
            list3.append(i)
print(list3)
```

```python
num = int(input("Enter number: "))
temp = num
rev = 0

while num > 0:
    rev = rev*10 + num%10
    num //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
num = int(input("Enter number: "))
temp = num
sum = 0
n = len(str(num))

while num > 0:
    d = num % 10
    sum += d ** n
    num //= 10

if temp == sum:
    print("Armstrong")
else:
    print("Not Armstrong")
```
